```python
#https://towardsdatascience.com/scraping-google-maps-reviews-in-python-2b153c655fc2

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Initialize webdriver
options = Options()
options.add_argument('--lang=en')
driver = webdriver.Chrome('/Users/Coding/Downloads/chromedriver')
```

```python
hvoe = 'https://www.google.com/maps/place/Hôpital+vétérinaire+pour+oiseaux+et+animaux+exotiques+de+Montréal+inc./@45.467644,-73.6229437,17z/data=!3m1!4b1!4m5!3m4!1s0x4cc910b434a1fa8f:0x8f2231a6fb040544!8m2!3d45.467644!4d-73.620755'
driver.get(hvoe)
```

```python
response = BeautifulSoup(driver.page_source, 'html.parser')
rlist = response.find_all('div', class_='mapsConsumerUiSubviewSectionReview__section-review-content')
rlist
```

```python
content = driver.page_source
```

```python
dates = []
names = []
reviews = []
ratings = []

soup = BeautifulSoup(content)
```

```python
for r in rlist:
    rel_date = r.find('span', class_='section-review-publish-date').text
    name = r.find('div', class_='section-review-title').find('span').text
    try:
        review_text = r.find('span', class_='section-review-text').text
    except Exception:
        review_text = None
    stars = r.find('span', class_='section-review-stars')['aria-label']
    dates.append(rel_date)
    names.append(name)
    reviews.append(review_text)
    ratings.append(stars)
```

```python
df_hvoe = pd.DataFrame({'Date' : dates,
                   'Name' : names,
                   'Review' : reviews,
                   'Rating' : ratings}
                 ) 
df_hvoe
```

```python
df_hvoe.to_csv('HVOE.csv', index=False, encoding='utf-8')
```
